# Graphs

The objective of this notebook is to take the CSV file that was created using various files and then attempt to build graphs comparing the number of threads to the speed of the process

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Smith-Waterman Graphs

```python
column_names = ['n', 'm', 'chunksize', 'runtime', 'threads']
df_sw = pd.read_csv('Smith-Waterman/sw.csv', names=column_names)
```

```python
df_filtered = df_sw[['n', 'runtime', 'threads']]

grouped_data = df_filtered.groupby('threads')

fig, ax = plt.subplots(figsize=(20, 12))  

for thread_count, group in grouped_data:
    group_sorted = group.sort_values('n')
    ax.plot(group_sorted['n'], group_sorted['runtime'], marker='o', label=f'Threads: {thread_count}')


ax.set_xlabel('Number of Elements in the Sequence')
ax.set_ylabel('Runtime in seconds')
ax.set_title('Runtime vs n for Different Thread Counts for Smith-Waterman')
ax.legend()

plt.show()

```

```python
df_new_sw = df_sw[df_sw['n'] <= 4000]


df_filtered = df_new_sw[['n', 'runtime', 'threads']]

grouped_data = df_filtered.groupby('threads')

fig, ax = plt.subplots(figsize=(20, 12))  

for thread_count, group in grouped_data:
    group_sorted = group.sort_values('n')
    ax.plot(group_sorted['n'], group_sorted['runtime'], marker='o', label=f'Threads: {thread_count}')


ax.set_xlabel('Number of Elements in the Sequence')
ax.set_ylabel('Runtime in seconds')
ax.set_title('Runtime vs n for Different Thread Counts for Smith-Waterman')
ax.legend()

plt.show()
```

```python
df_new_sw = df_sw[df_sw['n'] <= 4000]


df_filtered = df_new_sw[['n', 'runtime', 'threads']]

grouped_data = df_filtered.groupby('threads')

fig, ax = plt.subplots(figsize=(20, 12))  

for thread_count, group in grouped_data:
    group_sorted = group.sort_values('n')
    ax.plot(group_sorted['n'], group_sorted['runtime'], marker='o', label=f'Threads: {thread_count}')

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('Number of Elements in the Sequence')
ax.set_ylabel('Runtime in seconds')
ax.set_title('Runtime vs n for Different Thread Counts for Smith-Waterman')
ax.legend()

plt.show()
```

```python
df_filtered = df_sw[['chunksize', 'runtime', 'threads']]

grouped_data = df_filtered.groupby('threads')
fig, ax = plt.subplots(figsize=(20, 12)) 

for thread_count, group in grouped_data:
    group_sorted = group.sort_values('chunksize')
    ax.plot(group_sorted['chunksize'], group_sorted['runtime'], marker='o', label=f'Threads: {thread_count}')

ax.set_xlabel('Chunksize')
ax.set_ylabel('Runtime in seconds')
ax.set_title('Runtime vs Chunksize for Different Thread Counts for Smith-Waterman')
ax.legend()

plt.show()
```
